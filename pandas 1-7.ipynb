# Filtering dataframe on a condition

```python
data['CITY']=="Faisalabad"
```

```python
type(data['CITY']=="Faisalabad")
```

```python
mask=data['CITY']=="Faisalabad"                      # show only fsd students
data[mask]
```

```python
data[mask].shape
```

```python
def get_CITY_NAME_count(CITY):
   mask=data['CITY']==CITY           # show only LAHORE students
   return data[mask]
get_CITY_NAME_count('Lahore')
```

```python
mask1=data['CITY']=="Faisalabad"    
mask2=data['PERCENTAGE']<"50%"
data[mask1 & mask2].shape[0]
```

```python
mask1=data['CITY']=="Layyah"    
mask2=data['PERCENTAGE']<"80%"
data[mask1 & mask2]
```
